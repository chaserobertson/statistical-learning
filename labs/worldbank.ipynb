# Welcome

```r
library(tidyverse)
```

## Import the dataset

```r
esgdata = read.csv("~/Downloads/ESG_csv/ESGData.csv")
```

## Check out the columns

```r
names(esgdata)
```

## Ew, collapse all those nasty year columns into one

```r
years_collapsed <- esgdata %>%
    select(-Country.Code, -Indicator.Code) %>%
    pivot_longer(-c(Country.Name, Indicator.Name), 'year')

head(years_collapsed)
```

## Now get an idea of which indicators have the most data available(not missing)

```r
count_datapoints <- years_collapsed %>%
    group_by(Indicator.Name) %>%
    summarise(n_dp = sum(is.na(value))) %>%
    arrange(desc(n_dp))

count_datapoints
```
